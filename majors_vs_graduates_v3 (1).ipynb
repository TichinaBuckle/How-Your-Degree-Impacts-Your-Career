# Conclusions / Summary

### Conclusions about the All (All Graduate Who have done a Major)

* The average annual salary of a graduate was found to be $57k.

* Engineering Majors are by far the most lucrative category of degree, enjoying a median annual salary of $78k.

* Art and Interdisciplinary Majors made up the least lucrative major, with an art major expecting a median salary of 43k and an interdisciplinary major expecting a median salary of 43k post-graduation.

* Unsurprising most of the best paying majors 8/10 were Engineering majors, with the highest paying major being Petroleum Engineering with a median salary of 125k.

* While the overall lowest-earning major from the dataset was Neuroscience with a median salary of 35k per year.

* The average unemployment of a graduate was found to be 6%.

* Agriculture and Natural Resources majors held the lowest unemployment rate of 3.9%, followed by Education and Health with unemployment rates of 4.7%.

* Art and Interdisciplinary majors while making up the least lucrative major, also had the highest unemployment rates with Art majors expecting an unemployment rate of 8.8% and an interdisciplinary major expecting an unemployment rate of 7.7% post-graduation.

### Conclusions about the Graduate Level Majors

* Engineering still made up most of the highest-earning degrees, dropping from 8/10 to 6/10.

* Heath and Medical Preparatory Programs were the overall highest-earning majors at the graduate level, with a median salary of 135k.

* Cosmetology Services and Culinary Arts were the overall lowest-earning majors at the graduate level, with a median salary of 47k.

* Engineering majors remained the category of degrees with the highest earning graduates even at the graduate level, with the median salary jumping 94k, and the gap between the top majors closing.

* The second place was Physical Sciences with median salaries of 89k.

* The third Place was Computers & Mathematics with median salaries of 85k.

* The lowest-earning majors at the graduate level were Interdisciplinary, Arts, and Education. Interdisciplinary earning a median salary of 55k, Arts majors earning 57k and Education earning 58k. However, Interdisciplinary and Education majors experience a low unemployment rate, with Interdisciplinary having the lowest unemployment rate of 2% and Education having an unemployment rate of 2.7%.

* The category of major with the biggest jump from all to graduate level was Biology and Life Science with a 32k increase and the category with the lowest jump was Interdisciplinary was an increase of 12k.

### Conclusions about Women with STEM Majors

* The women in STEM make on average a median salary of 43k annually, a median salary of 20k less than the general pollution (of STEM majors).

* The highest earning major for women mostly remained the same as the general population.

* The category with the large of share women was Health at 80%, while Engineering had by far the lowest at 24% women, likely playing some factor in the significant disparity between the median annual salary.
